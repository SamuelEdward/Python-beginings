```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_data_points = 10

my_figure = plt.figure()
subplot_1 = my_figure.add_subplot(1, 1, 1)
subplot_1.plot(np.random.random(number_of_data_points).cumsum())

subplot_1.text (1, 0.5, 'An equation: $E-mc^2$', fontsize=18, color='red')
subplot_1.text(1, 1.5, 'Hello, Moutain climbing!', family='monospace', fontsize=14, color='green')

subplot_1.text(0.5, 0.5, 'We are centered now', transform=subplot_1.transAxes)

subplot_1.annotate('shoot arrow', xy=(2,1), xytext=(3,4),
                  arrowprops=dict(facecolor='red', shrink=0.05))

plt.show()
```
